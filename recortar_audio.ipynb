```python
from pydub import AudioSegment
import numpy as np
import os

def dividir_audio_aleatorio(audio_path, duracao_minima_segmento, duracao_maxima_segmento):
    # Carregar o arquivo de áudio
    audio = AudioSegment.from_file(audio_path)

    # Obter a duração total do áudio
    duracao_total = len(audio)

    # Lista para armazenar os áudios divididos
    audios_divididos = []

    # Realizar divisões aleatórias
    inicio = 0
    while inicio < duracao_total:
        # Gerar a duração aleatória do próximo segmento
        duracao_segmento = np.random.uniform(duracao_minima_segmento, duracao_maxima_segmento)

        # Verificar se o final do segmento ultrapassa a duração total
        fim_segmento = min(inicio + duracao_segmento, duracao_total)

        # Dividir o áudio
        segmento = audio[inicio:fim_segmento]

        # Adicionar o segmento à lista
        audios_divididos.append(segmento)

        # Atualizar o ponto de início para o próximo segmento
        inicio = fim_segmento

    return audios_divididos

# Pasta de entrada (áudios originais)
pasta_audio = "caminho para pasta de audios"

# Pasta de saída (áudios divididos)
pasta_audios_divididos = "caminho para pasta de audios divididos"

# Certifique-se de que a pasta de saída exista, se não, crie-a
if not os.path.exists(pasta_audios_divididos):
    os.makedirs(pasta_audios_divididos)

# Definir a duração mínima e máxima do segmento em milissegundos
duracao_minima_segmento = 1000  # Exemplo: 1 segundo
duracao_maxima_segmento = 5000  # Exemplo: 5 segundos

# Iterar sobre os arquivos na pasta "audio"
for nome_arquivo in os.listdir(pasta_audio):
    caminho_audio = os.path.join(pasta_audio, nome_arquivo)

    # Verificar se o arquivo é um arquivo de áudio
    if nome_arquivo.lower().endswith((".mp3", ".wav")):
        print(f"Dividindo áudio: {nome_arquivo}")
        
        # Gerar a lista de áudios divididos
        audios_divididos = dividir_audio_aleatorio(caminho_audio, duracao_minima_segmento, duracao_maxima_segmento)

        # Salvar áudios divididos na pasta "audios_divididos"
        for i, audio_dividido in enumerate(audios_divididos):
            nome_audio_dividido = f"{os.path.splitext(nome_arquivo)[0]}_dividido_{i + 1}.wav"
            caminho_audio_dividido = os.path.join(pasta_audios_divididos, nome_audio_dividido)
            audio_dividido.export(caminho_audio_dividido, format="wav")
            print(f"Áudio dividido salvo em: {caminho_audio_dividido}")
```
